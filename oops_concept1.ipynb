```python
class Pattern():
  n = int(input("Enter a value:"))

  def raw(self):
    for r in range(1,self.n+1):
      for c in range(1,r+1):
        print("* ",end = '')
      print()

  def fr(self):
    for r in range(1,self.n+1):
      for s in range(0,self.n-r):
        print(" ",end = '')
      for c in range(1,r+1):
        print("*",end = '')
      print()


a = Pattern()
a.raw()
a.fr()
```
